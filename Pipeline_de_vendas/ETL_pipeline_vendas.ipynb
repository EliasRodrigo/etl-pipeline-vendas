```python
import pandas as pd
import numpy as np
```

```python
import os

os.makedirs("data/raw", exist_ok=True)
os.makedirs("data/processed", exist_ok=True)
```

```python
data = {
    "date": [
        "2024-01-01","2024-01-02","2024-01-03",
        "2024-01-04","2024-01-05","2024-01-06"
    ],
    "product": [
        "Notebook","Mouse","Teclado",
        "Notebook","Mouse","Monitor"
    ],
    "category": [
        "Eletronicos","Eletronicos","Eletronicos",
        "Eletronicos","Eletronicos","Eletronicos"
    ],
    "price": [3500, 80, 150, 3400, 90, 1200],
    "quantity": [2,5,3,1,6,2],
    "city": [
        "São Paulo","Rio de Janeiro","Belo Horizonte",
        "São Paulo","Curitiba","Rio de Janeiro"
    ]
}
```

```python
# Isso cria a tabela.

df = pd.DataFrame(data)

df
```

```python
df.to_csv("data/raw/vendas.csv", index=False)
```

```python
# Isso é o E do ETL (Extract).

df_raw = pd.read_csv("data/raw/vendas.csv")

df_raw.head()
```

```python
# Isso mostra: tipos das colunas, quantidade de valores, se existem dados faltantes

df_raw.info()
```

```python
# Isso mostra estatísticas das colunas numéricas:
# média, mínimo, máximo, desvio padrão
df_raw.describe()
```

```python
# Isso mostra quantos valores faltando existem em cada coluna.
df_raw.isnull().sum()
```

```python
# Isso converte a coluna para datetime

df = df_raw.copy()

df["date"] = pd.to_datetime(df["date"])

df.head()
```

```python
# faturamento = price * quantity

df["revenue"] = df["price"] * df["quantity"]

df.head()
```

```python
# Criar colunas de tempo , ano, mês e dia

df["year"] = df["date"].dt.year
df["month"] = df["date"].dt.month
df["day"] = df["date"].dt.day

df.head()
```

```python
df.to_csv("data/processed/vendas_tratadas.csv", index=False)
```
